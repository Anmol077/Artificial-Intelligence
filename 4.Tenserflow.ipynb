```
import numpy as np
import tensorflow as tf
```

```
hello = tf.constant("Hello World")
print(hello)
```

```
#define tf session
sess = tf.Session()

#run the session
print(sess.run(hello))
```

```
#Rank 0 tensor
tensor = tf.constant(2)
shape = tf.shape(tensor)
rank = tf.rank(tensor)
print(tensor)
```

```
sess.run(rank)
```

```
sess.run(shape)
```

```
tensor = tf.constant([1,2,3])
shape = tf.shape(tensor)
rank = tf.rank(tensor)
print(tensor)
print(sess.run(rank))
sess.run(shape)
```

```
tensor = tf.constant([[1,2,3],[4,5,6]])
shape = tf.shape(tensor)
rank = tf.rank(tensor)
print(tensor)
print(sess.run(rank))
sess.run(shape)
```

```
tensor = tf.constant([[[1,2,3],[4,5,6]]])
shape = tf.shape(tensor)
rank = tf.rank(tensor)
print(tensor)
print(sess.run(rank))
sess.run(shape)
```

```
a = tf.constant([1.0,2.0],dtype = tf.float32)
b = tf.constant([3.0,4.0])
total = a+b
print(a)
print(b)
print(total)
```

```
sess = tf.Session()
sess.run(total)
```

```
with tf.Session():
  x = tf.constant([[2,4],[5,6]])
  y = tf.constant([[1,0],[0,1]])
  z = tf.matmul(x,x) + tf.matmul(x,y)
  #run graph
  result = z.eval() #short for sess.run(z)
```

```
result
```

```
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)

z = x+y
# Define session for executing the graph
# Feed_dict argument of the run method to feed concrete values
print(sess.run(z, feed_dict = {x:4, y:5.2}))
print(sess.run(z, feed_dict ={x: [1,2], y:[5,6]}))
```
