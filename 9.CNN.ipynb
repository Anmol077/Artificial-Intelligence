```
import tensorflow as tf
```

```
(x_train, y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```
import matplotlib.pyplot as plt
image_index = 777 #select of image
print(y_train[image_index])
plt.imshow(x_train[image_index],cmap = "Greys")# for black and white
```

```
x_train.shape
```

```
#reshapeing the array to 4 dims so that it can work with keras API
x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)
input_shape = (28,28,1)
#Making sure that the value are float so that we ge decimal points after division
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
#Normalizing  the RGB codes by dividing it to the max RGB value.
x_train /=255
x_test /= 255
print(('x_train shape',x_train.shape))
print("Number of images in x_train",x_train.shape[0])
print("Number of image in x_test",x_test.shape[0])
```

```
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPool2D
```

```

model = Sequential()
model.add(Conv2D(28,kernel_size=(3,3),input_shape=input_shape))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(10,activation=tf.nn.softmax))
```

```
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x=x_train,y=y_train,epochs=10)
```

```
#evaluating the model
model.evaluate(x_test,y_test)
```

```
image_index = 4443
plt.imshow(x_test[image_index].reshape(28,28),cmap = "Greys")# for black and white
pred = model.predict(x_test[image_index].reshape(1,28,28,1))
print(pred.argmax())
```
